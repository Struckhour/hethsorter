```python
import os
import shutil

path = 'HERMIT4_20220506_072428-s' + '/'

for slicefolder in os.listdir(path):
    if 'chunks' not in slicefolder and os.path.isdir(path + slicefolder):
        #songs
        for imgfile in os.listdir(path + slicefolder + '/' + 'training_songs/positives/'):
            if not os.path.exists('../tensorthrush/songs/pos/' + imgfile):
                shutil.copy(path + slicefolder + '/' + 'training_songs/positives/' + imgfile, '../tensorthrush/songs/pos/' + imgfile)
        for imgfile in os.listdir(path + slicefolder + '/' + 'training_songs/negatives/'):
            if not os.path.exists('../tensorthrush/songs/neg/' + imgfile):
                shutil.copy(path + slicefolder + '/' + 'training_songs/negatives/' + imgfile, '../tensorthrush/songs/neg/' + imgfile)
        #intros
        for imgfile in os.listdir(path + slicefolder + '/' + 'training_intros/positives/'):
            if not os.path.exists('../tensorthrush/intros/pos/' + imgfile):
                shutil.copy(path + slicefolder + '/' + 'training_intros/positives/' + imgfile, '../tensorthrush/intros/pos/' + imgfile)
        for imgfile in os.listdir(path + slicefolder + '/' + 'training_intros/negatives/'):
            if not os.path.exists('../tensorthrush/intros/neg/' + imgfile):
                shutil.copy(path + slicefolder + '/' + 'training_intros/negatives/' + imgfile, '../tensorthrush/intros/neg/' + imgfile)
                
        #error songs
        if os.path.exists(path + slicefolder + '/errors'):
            for imgfile in os.listdir(path + slicefolder + '/errors/training_songs/positives/'):
                if not os.path.exists('../hermiterrors/songs/pos/' + imgfile):
                    shutil.copy(path + slicefolder + '/errors/training_songs/positives/' + imgfile, '../hermiterrors/songs/pos/' + imgfile)
            for imgfile in os.listdir(path + slicefolder + '/errors/training_songs/negatives/'):
                if not os.path.exists('../hermiterrors/songs/neg/' + imgfile):
                    shutil.copy(path + slicefolder + '/errors/training_songs/negatives/' + imgfile, '../hermiterrors/songs/neg/' + imgfile)
        #error intros
            for imgfile in os.listdir(path + slicefolder + '/errors/training_intros/positives/'):
                if not os.path.exists('../hermiterrors/intros/pos/' + imgfile):
                    shutil.copy(path + slicefolder + '/errors/training_intros/positives/' + imgfile, '../hermiterrors/intros/pos/' + imgfile)
            for imgfile in os.listdir(path + slicefolder + '/errors/training_intros/negatives/'):
                if not os.path.exists('../hermiterrors/intros/neg/' + imgfile):
                    shutil.copy(path + slicefolder + '/errors/training_intros/negatives/' + imgfile, '../hermiterrors/intros/neg/' + imgfile)
                
```
